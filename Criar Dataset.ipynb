```python
# pega despesas
years = list(range(2015, 2019))
for year in years:
    description = "Apresentam  informações relativas aosdados cadastrais e valoresdos instrumentos contratuais cadastrados no  sistema corporativo da PCRJ (Sistema de Controle de Contratos –FCTR) de {}.".format(year).replace("  ", " ")
    url = "http://riotransparente.rio/riotransparente/arquivos/Open_Data_Contratos_{}.txt".format(year)
    manager.create_dataset("riotransparente_contratos_{}".format(year), url, description, local_source="./data/Open_Data_Contratos_{}.csv".format(year), format="csv")
```
