```python
# pega despesas
years = list(range(2015, 2019))
for year in years:
    description = "Apresentam informações relativas a cada ato de execução orçamentária e de restos a pagar realizados no ano selecionado. Esta consulta se diferencia da consulta “Favorecidos” por apresentar os atos de realização (e cancelamentos)  de  empenhos,  com  seus  respectivos  valores,  que  ainda  não  tenham  sido liquidados,  além  de  campos  indicativos  de  ação  com serviço  público  de  saúde  ou  de manutenção e desenvolvimento de ensino. Cada registro da consulta corresponde a um ato de realização ou de cancelamento de {}.".format(year).replace("  ", " ")
    url = "http://riotransparente.rio/riotransparente/arquivos/Open_Data_Desp_Ato_{}.txt".format(year)
    manager.create_dataset("riotransparente_despesas_ato_{}".format(year), url, description, local_source="./data/Open_Data_Desp_Ato_{}.csv".format(year), format="csv")
```
